```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the causal edges leading to the four outputs
edges = [
    ("X20", "VX"), ("X27", "VX"), ("X73", "VX"), ("X96", "VX"),
    ("X30", "VY"), ("X59", "VY"), ("X60", "VY"), ("X80", "VY"), ("X99", "VY"), ("X116", "VY"), ("X120", "VY"),
    ("X30", "VZ"), ("X43", "VZ"), ("X73", "VZ"), ("X75", "VZ"), ("X100", "VZ"), ("X125", "VZ"),
    ("X34", "omega_z"), ("X50", "omega_z"), ("X69", "omega_z"), ("X72", "omega_z"), ("X81", "omega_z"), ("X84", "omega_z")
]

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Use a layout that spreads out nodes better
pos = nx.kamada_kawai_layout(G)  # Kamada-Kawai layout helps reduce overlap

plt.figure(figsize=(14, 12))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=1500, node_color='skyblue')

# Draw edges with clearer arrows
nx.draw_networkx_edges(
    G, pos,
    arrowstyle='->',
    arrowsize=50,
    edge_color='black',
    width=2
)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

plt.title("Causal Graph to VX, VY, VZ, omega_z", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()

```
